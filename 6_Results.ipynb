```python
df = pd.read_csv('model_info.txt', names=["Model Name", "RMSE", "R2_score"])
df = df.drop_duplicates(subset=["Model Name"])
df = df.reset_index(drop=True)
df.sort_values(by="RMSE", ascending=False, inplace=True)
```

```python
df.head(10)
```

```python
plt.figure(figsize=(12, 8))
bars = plt.bar(df["Model Name"], df["RMSE"], color='skyblue')

plt.xlabel('Model Name', fontsize=14)
plt.ylabel('RMSE', fontsize=14)
plt.title('Model RMSE Comparison', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# Add values on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.01, f'{yval:.2f}', ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
bars = plt.bar(df["Model Name"], df["R2_score"], color='orange')

plt.xlabel('Model Name', fontsize=14)
plt.ylabel('R2_score', fontsize=14)
plt.title('Model R2_score Comparison', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)


for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.01, f'{yval:.2f}', ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```

### Conclusion

The performance of models: Linear Regression, Ridge Regression, Random Forest, Support Vector Machine (SVM) and Neural Network, was evaluated using Root Mean Squared Error (RMSE) and R-squared (R²) as the key metrics.

As a baseline, the mean of the target variable (pIC50) was used for comparison. This provides a reference point to assess whether the models offer meaningful improvements over a naïve prediction strategy.

Linear and Ridge Regression showing moderate predictive ability but limited capacity to model complex, non-linear relationships in the data.

SVM achieved the best overall performance, with the lowest RMSE(0.76) and highest R²(0.71), demonstrating strong generalization and the ability to capture non-linear patterns effectively.

Random Forest also delivered competitive results, particularly when hyperparameters and kernels were well-tuned, performing similarly to SVM in some cases.

Neural Networks showed promising results as well, benefiting from their ability to model complex, high-dimensional relationships. While their performance was comparable to Random Forest and SVM in terms of RMSE and R², neural networks require careful tuning and more computational resources. They offer strong predictive power but may lack interpretability compared to tree-based or linear models.

Overall, the results indicate that non-linear models like Random Forest and SVM are better suited for predicting kinase bioactivity due to the intricate nature of molecular interactions. However, linear models remain valuable for their simplicity, speed, and interpretability. 

In conclusion, this study demonstrates the value of machine learning in predicting kinase inhibitor bioactivity and provides a foundation for further optimization and integration into virtual screening pipelines for early-stage drug discovery.
